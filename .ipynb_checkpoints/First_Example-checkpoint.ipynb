## 第一个机器学习样例

根据47个房子的面积和价格，建立模型预测新的房价（随面积的变化）。

### 一、获取与处理数据

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# 输入数据
x, y = [], []

# 按行读取文件，r为读取操作
for sample in open("../../MachineLearning/_Data/prices.txt", 'r'):
    _x, _y = sample.split(',') # 以逗号分隔
    x.append(float(_x)) # 转换为浮点数
    y.append(float(_y))

# 转为ndarray格式，方便向量运算
# 数据个数确定，不需要增删，且需要随机读取，用数组格式储存合理！
x, y = np.array(x), np.array(y)

# 通常对输入数据进行标准化，x为房子面积（自变量），y为房子价格
x = (x - x.mean()) / x.std()

# 画散点图
plt.figure()
plt.scatter(x, y, c='g', s=6)
plt.show()
```

### 二、选择与训练模型

观察图像，决定使用线性回归中的多项式拟合。

```python
# 在（-2，4）区间均匀取100个点
x0 = np.linspace(-2, 4, 100)

# 多项式回归模型
# deg参数代表模型参数中的n，即多项式的次数
# 返回的模型能够通过输入的x，输出相应预测的y
def get_model(deg):
    return lambda input_x=x0 : np.polyval(np.polyfit(x, y, deg), input_x)

# 注意上述的lambda表示输出一个新建立的函数，默认参数为x0，也可自行输入x值，便可输出相应的y值
```

* polyfit函数：该函数返回多项式的最小参数p（数组），即多项式各项系数——该函数就是模型的训练函数
* polyval函数：根据多项式的各项系数p、多项式中x的值，返回相应预测的y值

### 三、评估和可视化结果

我们训练的目的是使得损失函数最小化，所以用损失函数来衡量模型好坏似乎很合理。

```python
# 根据参数n、输入的x、y返回相应的损失
# 根据P15的损失函数定义
def get_cost(deg, input_x, input_y):
    return 0.5 * ((get_model(deg)(input_x) - input_y) ** 2).sum()

# 定义测试集，deg的不同取值
test_set = (1, 4, 10)
for d in test_set:
    print(get_cost(d, x, y))
```

发现（1，4，10）对应的损失函数值逐渐减小，似乎deg=10是最好的选择，而deg=1最差。

用画图的方式最方便观察是否过拟合。

```python
# 画出原始数据散点图
plt.scatter(x, y, c='g', s=20)
for d in test_set:
    plt.plot(x0, get_model(d)(), label='degree = %d' % (d)) 
    # label='degree = {}'.format(d) 格式输出的两种表示

# 设置横轴纵轴的范围
plt.xlim(-2, 4)
plt.ylim(1e5, 8e5)

# 显示label信息
plt.legend()
plt.show()
```

发现deg=4已经出现过拟合现象，所以最好的选择应该是deg=1。
